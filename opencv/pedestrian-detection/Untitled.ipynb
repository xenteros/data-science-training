```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
image_path = 'images/person2.png'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# detect people in the image
(rectangles, weights) = hog.detectMultiScale(image, winStride=(4, 4), padding=(3, 3), scale=1.05)
```

```python
for (x, y, w, h) in rectangles:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)
```

Zastosować metodę "Non Maximum Suppression" w odniesieniu do ograniczników, stosując dość duży próg nakładania się, aby spróbować utrzymać nakładające się skrzynie, które nadal są ludźmi.

```python
rectangless = np.array([[x, y, x + w, y + h] for (x, y, w, h) in rectangles])
```

```python
for (xA, yA, xB, yB) in rectangless:
    cv2.rectangle(image, (xA, yA), (xB, yB), (0, 255, 0), 2)
    
plt.imshow(image)
```
