#### Funkcja tworząca wykres

```python
def plot_timeseries(axis, x, y, color, xlabel, ylabel):
    axis.plot(x, y, color=color)
    axis.set_xlabel(xlabel)
    axis.set_ylabel(ylabel, color=color)
    axis.tick_params('y', colors=color)
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, merged.index, merged['Interpolated'],
                'r', 'Time', 'CO2')
ax2 = ax.twinx()
plot_timeseries(ax2, merged.index, merged['AverageTemperature'],
                'b', 'Time', 'Temperature')
```

#### Z analizą danych

```python
medals = pd.read_csv('datasets/olympic.csv', index_col='Year')
```

```python
medals.head()
```

```python
def groupby_column(dataframe, column):
    groupby = dataframe.groupby(column)
    return [groupby.get_group(x) for x in groupby.groups]
```

```python
def analyze_medals(medal_df):
    counts = medal_df["Medal"].value_counts()
    gold = counts['Gold'] if 'Gold' in counts else 0
    silver = counts['Silver'] if 'Silver' in counts else 0
    bronze = counts['Bronze'] if 'Bronze' in counts else 0
    return medal_df['NOC'].iloc[0], gold, silver, bronze
```

```python
medals_per_country = {}
for country_df in groupby_column(medals, "NOC"):
    country, gold, silver, bronze = analyze_medals(country_df)
    medals_per_country[country] = [gold, silver, bronze]
new_medals = pd.DataFrame.from_dict(medals_per_country,
                                    orient='index',
                                    columns=['Gold', 'Silver', 'Bronze'])
new_medals
```

```python
new_medals = new_medals[new_medals.sum(axis=1) > 500]
new_medals
```

```python
# stacked barchart
fig, ax = plt.subplots()
ax.bar(new_medals.index, new_medals["Gold"], label="Gold")
ax.bar(new_medals.index, new_medals["Silver"],
       bottom=new_medals["Gold"], label="Silver")
ax.bar(new_medals.index, new_medals["Bronze"],
       bottom=new_medals["Gold"]+new_medals["Silver"], label="Bronze")
ax.xaxis.set_tick_params(rotation=90)
ax.set_ylabel('Medals')

ax.legend()

plt.show()
```

## Histogramy

Do wyświetlania nie tylko liczności, ale też rozkładu wartości

```python
all_sports = {}
for sport_df in groupby_column(medals, 'Sport'):
    all_sports[sport_df['Sport'].iloc[0]] = sport_df
all_sports.keys()
```
