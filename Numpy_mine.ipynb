```python
def fib(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a+b
    return a

%timeit fib(8)
```

```python
import numpy as np

def n_fib(n):
    A = np.array([0,1])
    for i in range(n):
        A = [A[1], A[0]+A[1]]
    return A[0]

%timeit n_fib(8)
```

```python
def fib(n):
    if n == 0 or n == 1:
        return n
    else:
        return fib(n-1)+fib(n-2)
    
%timeit fib(8)
```

```python
def fib(n):
    if n == 0: return 0

    n -= 2

    ca, cb, cc = sa, sb, sc = 1, 1, 0

    while n>0:
        if n & 1: 
            sa, sb, sc = (sa*ca + sb*cb), (sa*cb + sb*cc), (sb*cb + sc*cc)
        n >>= 1
        t = cb**2
        ca, cb, cc = (ca**2+t), (cb*(ca+cc)), (cc**2+t)

    return sa

%timeit fib(8)
```
